```python
pres =  {'Name' : ['Barack','George','Bill','George','Ronald','Jimmy','Gerald','Richard','Lyndon','John','Dwight','Harry','Franklin'],
         'LName': ['Obama', 'Bush', 'Clinton','Bush','Reagan','Carter','Ford','Nixon','Johnson','Kennedy','Eisenhower','Truman','Roosevelt'],
         'StartYear' : [2009,2001,1993,1989,1981,1977,1974,1969,1963,1961,1953,1945,1933],
         'EndYear' : [2015,2009,2001,1993,1989,1981,1977,1974,1969,1963,1961,1953,1945]
}

names = pd.DataFrame(pres)
```

```python
def nameCounter():
    
    # define empty lists
    fnames = []
    lnames = []
    counts = []
    lcounts = []
    totals = []
    years = []
    
    # loop through the names dictionary
    for index, row in names.iterrows():
        
        # loop through each year of each individual's Presidency
        for year in range(row['StartYear']-1,row['EndYear']+1):
            
            # create a subset data frame for the particular year you are on
            y = allStates[(allStates['Year']==year)] 
            
            # count the occurence of names
            count = ((sum(y['Count'][y['Name']==row['Name']])))   # first name
            lcount = ((sum(y['Count'][y['Name']==row['LName']]))) # last name
            total = ((sum(y['Count'])))                           # total names
            
            # add items to lists above
            fnames.append(row['Name'])
            lnames.append(row['LName'])
            counts.append(count)
            lcounts.append(lcount)
            totals.append(total)
            years.append(year)
            
        # create a data frame    
        final = pd.DataFrame({'f_name': fnames,'l_name': lnames,'f_count': counts,'l_count':lcounts,'total':totals,'year':years})
    return final
```
